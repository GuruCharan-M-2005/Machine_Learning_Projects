# Customer Group Classification

*Customer groups are classified as basic, expected and excited customer demands. Customers put forward basic demands for enterprise's product (or service). For customers, these demands are properties or functions that products (or services) must possess.*



### We are going to use K-means Clustering

## Importing Libraries


```python
import pandas as pd
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Mall_Customers.csv')
df=pd.DataFrame(data)
df.head()
```

## Data Pre-Processing

```python
df=df.drop(['CustomerID'],axis='columns')
df.head()
```

```python
le=LabelEncoder()
df['Gender']=le.fit_transform(df.Gender)
df.head()
```

## Data Grouping

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(df)
kmeans
```

```python
df['labels']=kmeans.labels_
df.head()
```

```python
df.tail()
```

### Therefore they are grouped into 3
# Visualisation

```python
fig=px.histogram(df,x='labels')
fig.update_layout(bargap=0.2)
```

```python
px.scatter(df,x='Age',y='Annual Income (k$)')
```
